# APPLY WEB SCRAPER

```python
from bs4 import BeautifulSoup
from time import sleep
```

```python
def get_url(artist_url, song_name_url, is_mobile = True):
    prefix = "m." if is_mobile else ""
    song_url = "https://%scifraclub.com.br/%s/%s/#instrument=guitar" % (prefix ,artist_url, song_name_url)
    return song_url
```

```python
def get_chords(artist_url, song_name_url):
    #sleep(0.01)
    try:
        r = requests.get(get_url(artist_url, song_name_url), timeout=3)
    except:
        return []
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    chords = soup.find_all("div", {"class":"chordWrapper"})

    chords_list = []

    for i in range(len(chords)):
        chords_list.append(chords[i].find("strong").text)


    return chords_list
```

```python
df["chords"] = df.apply(lambda dr: get_chords(dr.url_artist, dr.url_song), axis=1)
df["url"] = df.apply(lambda dr: get_url(dr.url_artist, dr.url_song, False), axis=1)
df
```

```python
def encodeMod (mod):
    if mod == 'video aula':
        return 0
    elif mod == 'revisada':
        return 1
    else:
         return 2


df.drop(columns=["genre.url", "url_artist", "url_song"], inplace=True)

df.rename(columns={"name_song": "song", "name_artist": "artist", "genre.name": "genre"}, inplace=True)

df['mod'] = df['mod'].map(encodeMod)

df = df[df['chords'].map(lambda d: len(d)) > 0]

df = df.drop_duplicates(subset=["song", "artist"])

df
```
