> APPLY WEB SCRAPER

```python
from bs4 import BeautifulSoup
```

```python
def get_url(artist_url, song_name_url):
    song_url = "https://m.cifraclub.com.br/%s/%s/#instrument=guitar" % (artist_url, song_name_url)
    return song_url
```

```python
def get_chords(artist_url, song_name_url):
    r = requests.get(get_url("henrique-e-juliano", "cidade-vizinha"))
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    chords = soup.find_all("div", {"class":"chordWrapper"})

    chords_str = ""

    for i in range(len(chords)):
        chords_str += chords[i].find("strong").text
        chords_str += ";"

    return chords_str
```

```python
df["chords"] = df.apply(lambda dr: get_chords(dr.url_artist, dr.url_song), axis=1)
```
